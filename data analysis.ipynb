# Customer Reviews

```python
cust_reviews = pd.read_excel('customer_reviews.xlsx')
cust_reviews.head()
```

```python
cust_reviews.info()
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
sia = SentimentIntensityAnalyzer()
```

```python
sia.polarity_scores('WOW! This product is awesome and great and sugest other to buy now')
```

```python
def sentiment_review(review):
    sentiment = sia.polarity_scores(review)
    return sentiment['compound']
```

```python
cust_reviews['ReviewText'] = cust_reviews['ReviewText'].str.replace('  ',' ')
```

```python
cust_reviews['sentiment_score'] = cust_reviews['ReviewText'].apply(sentiment_review)
```

```python
cust_reviews
```

```python
def sentiment_catogary(score, rating):
    if score < -0.05:
        if rating < 3:
            return 'Negative'         #negative score and negative rating
        elif rating == 3:
            return 'Mixed negative'   #negative score and neutral rating
        else:
            return 'Mixed positive'   #negative score and high rating
    elif score > 0.05:
        if rating > 3:
            return 'Positive'         #positive score and high rating
        elif rating == 3:
            return 'Mixed positive'    #positive score and neutral rating
        else:
            return 'Mixed negative'
    else:
        if rating > 3:
            return 'Positive'           # neutral score and high rating
        elif rating == 3:
            return 'Neutral'            # neutral score and neutral rating 
        else:
            return 'Negative'            #neutral score and negative rating  
            
```

```python
cust_reviews['sentiment_category'] = cust_reviews.apply(lambda row: sentiment_catogary(row['sentiment_score'],row['Rating']),axis=1)
```
